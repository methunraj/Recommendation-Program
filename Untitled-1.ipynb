Importing the dependencies:


```python
import numpy as np  #Numpy is used for numpy arrays
import pandas as pd #Pandas is used for creating Panda Dataframe (Also known as structedframe) 
import difflib #Difflib is used for comparing the datas to get the required information
from sklearn.feature_extraction.text import TfidfVectorizer #TfidVectorizer is used to convert the textual data to Nummerical data 
from sklearn.metrics.pairwise import cosine_similarity #Cosine_similarity is used to find the similarity between the datas and gives us the similarity score. Using this score we can predict the next possible recommendation.
```

Data Collection and Pre Processing


```python
# loading the data from the CSV file in the folder
movies_data = pd.read_csv("D:\Smart Systems Masters\ML\Movie Recommendation System\movies.csv")
```

```python
#Printing the first five rows of the dataframe
movies_data.head()
```

```python
#To find the number rows and columns in the dataframe
movies_data.shape
```

```python
#select the required features for the system 

required_features = ['genres', 'keywords','original_title','tagline', 'cast','vote_average']
print("Selected Features for the Recommendation System : ", required_features)
```

```python
#Replacing the missing values with a null string 
for feature in required_features:
    movies_data[feature] = movies_data[feature].fillna('')


```

```python
#Combining all the selected features in a dataframe
features_data = movies_data['genres']+''+movies_data['keywords']+''+movies_data['original_title']+''+movies_data['tagline']+''+movies_data['cast']
print(features_data)

```

```python
#converting the text data to feature vectors(numerical data)

vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(features_data)
print(feature_vector)
```

To find the similarity between the data we will use the Cosine Similiaity


```python
#Getting the similarity Score 
similarity = cosine_similarity(feature_vector)
print(similarity)
similarity.shape
```

```python
#Getting User Input for Recommendation
movie_name = input("Enter the Movie Name")
```

```python
#Creating a list with all the movie names given in the data set
list_of_all_the_movies = movies_data['title'].tolist()
list_of_all_the_movies
```

```python
#find the close match for the movie given by the user

find_close_movie_names = difflib.get_close_matches(movie_name,list_of_all_the_movies)
print(find_close_movie_names)
```

```python
close_match = find_close_movie_names[0]
print(close_match)
```

```python
#finding the index value of the corresponding movie name searched by the user
index_match_name = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_match_name)
```

```python
#finding the similar movies
similarity_scores = list(enumerate(similarity[index_match_name])) 
similarity_scores
```

```python
#Sorting the movies based on similarity
sort_similar_movies = sorted(similarity_scores, key= lambda x:x[1], reverse=True)
sort_similar_movies
```

```python
#print the similar movies based on the index value
print("Movie Suggested for you: \n ")
i = 1 
for movie in sort_similar_movies:
    index = movie[0]
    title_from_index= movies_data[movies_data.index == index]['title'].values[0]
    if(i<20):
        print(i, '.',title_from_index)
        i+=1 
```
